## Filling NULL

In the Orders Dataframe the only column with NaNs detected is "total_paid"

```python
o_df.loc[o_df["total_paid"].isna()]
```

```python
nan_orders = o_df.loc[o_df["total_paid"].isna(), "order_id"].to_list()
```

From here the dataframe Orderlines is needed to see if we can find the correct "total paid" afterall

```python
ol_df
```

1. We detect the rows with the same number found in "id_order"

```python
nan_orders
```

```python
ol_df.loc[ol_df["id_order"].isin(nan_orders)]
```

```python
ol_needed_rows = ol_df.loc[ol_df["id_order"].isin(nan_orders)]
```

2. Now we calculate the needed values and save them into a list

```python
x = (ol_needed_rows.groupby("id_order")["unit_price"].sum()).values
```

```python
x
```

Here we finally replace the NaN with real values

```python
o_df.loc[o_df["total_paid"].isna(), "total_paid"] = x
```

```python
o_df.loc[o_df["total_paid"].isna()]
```
